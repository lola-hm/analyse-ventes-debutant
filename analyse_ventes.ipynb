```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ventes.csv")
df
```

```python
df.info()
```

```python
# convertir la colonne date
df["date"] = pd.to_datetime(df["date"])
```

```python
df.info()
```

```python
# créer la colonne chiffre d'affaire
df["chiffre_affaires"] = df["prix"] * df["quantite"]
df
```

```python
ca_total = df["chiffre_affaires"].sum()
ca_total
```

```python
print(f"Le chiffre d'affaire total sur cette période de données est de {ca_total} €.")
```

```python
# calcul de la quantité vendue par produit
ventes_produits = df.groupby("produit")["quantite"].sum()
ventes_produits
```

```python
# visualisation

ventes_produits.plot(kind="bar")
plt.title("Quantité vendue par produit")
plt.xlabel("Produit")
plt.ylabel("Quantité vendue")
plt.show()
```

```python
# calcul du CA par jour
ca_par_jour = df.groupby("date")["chiffre_affaires"].sum()
ca_par_jour
```

```python
# visualtion du CA par jour
ca_par_jour.plot(marker="o")
plt.title("Evolution du chiffre d'affaires")
plt.xlabel("Date")
plt.ylabel("Chiffre d'affaires (€)")
plt.show()
```
